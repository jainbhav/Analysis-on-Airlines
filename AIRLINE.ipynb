```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the data

```python
df=pd.read_excel('Copy of Data_Train.xlsx')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Filling missing values

```python
df['Route'].mode()
```

```python
df['Route']=df['Route'].fillna(df['Route'].mode()[0])
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops']=df['Total_Stops'].fillna(df['Total_Stops'].mode()[0])
```

```python
df.isnull().sum()
```

# Changing Datatype of Date and time

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df.replace({'non-stop': 0,'2 stops': 2,'1 stop': 1,'3 stops': 3, '4 stops': 4},inplace=True)
```

```python
df
```

# 1st Insights : How many flights with respect to their stopages

```python
df['Total_Stops'].value_counts()
```

```python
plt.title('No of flights stopage')
plt.hist(df['Total_Stops'],color='purple')
plt.xlabel('Stopages')
plt.ylabel('No of Flights')
plt.xticks(df['Total_Stops'].unique());
```

```python
plt.bar(df['Airline'],df['Total_Stops'])
plt.xlabel('Airlines')
plt.ylabel('Total stopages')
plt.xticks(rotation=90);
```

# 2nd Insights: What flight is cheaper or expensive

```python
p=df.groupby(['Airline'])['Price'].agg(['sum','max']).reset_index().sort_values(by='sum',ascending=False)
p
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.bar(p['Airline'],p['sum'],label='overall profit')
plt.xticks(rotation=90)
plt.legend()

plt.subplot(1,2,2)
plt.bar(p['Airline'],p['max'],label='expensive and cheap')
plt.xticks(rotation=90)
plt.legend();
```

# 5th Insights: On what month have maximum flights to take off

```python
plt.figure(figsize=(8,8))
month=list(df['Month_of_journey'])
airline=list(df['Airline'])
plt.scatter(airline,month)
plt.xticks(rotation=90)
plt.yticks(df['Month_of_journey'].unique())
plt.title('Max flights to take off ')
plt.xlabel('Airlines')
plt.ylabel('Months');
```
